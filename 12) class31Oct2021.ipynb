# Filter function

```python
l5= [1,2,3,4,5,6,7,7,8,9,9,0,0,0,45,67]
```

```python
#find all even numbers or odd numbers from l5
```

```python
def test6(data):
    if data % 2==0:
        return True
    else: 
        return False
```

```python
list(filter(test6,l5))                          #IN filter returing value shoul be true or else vlaue won't be printed
```

```python
#other way-
list(filter(lambda data : True if data %2 ==0 else False,l5))
```

```python
#or
g=lambda data : True if data %2 ==0 else False
list(filter(g,l5))
```
